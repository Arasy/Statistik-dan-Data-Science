```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import time
```

```python
url = "http://detik.com"
s = requests.session()
r = s.get(url)
```

```python
r.text
```

```python
soup = BS(r.text)
```

```python
fin = []

for link in soup.find_all('a'):
    if "finance.detik.com" in link.get('href'):
        fin.append(link)
```

```python
fin
```

```python
detik = []

for link in fin:
    try:
        html = s.get(link.get('href'))
        html_soup = BS(html.text)
        header = html_soup.find('h1').text
        author = html_soup.find('div',{'class':'detail__author'}).text
        date = html_soup.find('div',{'class':'detail__date'}).text
        content = html_soup.find('div',{'class':'detail__body-text itp_bodycontent'}).text
        detik.append([header,author,date,content])
    except Exception as e:
        print("%s tidak bisa diakses" %link)
```

```python
detik
```

```python
df_detik = pd.DataFrame(detik)
```
