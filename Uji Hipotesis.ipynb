```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

Contoh kasus :

ada data berupa usia dari 20 orang sampel acak. kita ingin menguji apakah rata-rata usia dari populasinya 30 dengan nilai signifikansi 0.05

H0 = rata-rata usia adalah 30 tahun

```python
#One sampled t-test : menguji satu grup sampel

umur = np.genfromtxt("dataumur.csv",skip_header=1)
print(umur)
```

```python
umur_mean = np.mean(umur)
print(umur_mean)
```

```python
tset, pval = stats.ttest_1samp(umur,30)
```

```python
print("p-values = ",pval)
if pval < 0.05:
    print("H0 ditolak")
else :
    print("H0 diterima")
```

Kasus kedua, ada 2 grup sampel. satu diambil dari laki-laki (data_A), satu diambil dari perempuan (data_B). kita ingin menguji apakah tinggi laki-laki dan perempuan bisa dianggap sama atau berbeda secara signifikan.

H0 : tinggi laki-laki = tinggi perempuan

```python
#generate data

data_A = np.random.normal(170,10,10000)
data_B = np.random.normal(165,10,10000)
```

```python
plt.figure(figsize=(20, 10))
sns.kdeplot(data_A, label="Laki-Laki", shade = True, color = '#0000ff')
sns.kdeplot(data_B, label="Perempuan", shade = True, color = '#ff0000')
```

```python
#Two sampled t-test : membandingkan 2 sampel independen
tset,pval = stats.ttest_ind(data_A,data_B)
```

```python
#jika pvalue >= alpha, maka H0 diterima, sebaliknya, H0 ditolak
print("p-values = ",pval)
if pval < 0.05:
    print("H0 ditolak")
else :
    print("H0 diterima")
```
